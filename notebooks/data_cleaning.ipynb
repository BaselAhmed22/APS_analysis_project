```python
numeric_cols = new_df.select_dtypes(include=['number']).columns # List of numeric columns

outliers_summary = {} # Dictionary to store outliers count for each numeric column


for col in numeric_cols:
    Q1 = new_df[col].quantile(0.25)   # Q1
    Q3 = new_df[col].quantile(0.75)   # Q3
    IQR = Q3 - Q1                 # Interquartile Range (IQR)

    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR

    outliers = new_df[(new_df[col] < lower_bound) | (new_df[col] > upper_bound)]   # Identify outliers

    outliers_summary[col] = len(outliers) # Count of outliers in each numeric column

# Create a DataFrame to display the outliers summary
outliers_df = pd.DataFrame(list(outliers_summary.items()), columns=['Column', 'Outlier_Count'])
outliers_df = outliers_df.sort_values(by='Outlier_Count', ascending=False)

display(outliers_df)
```

```python
# relationship between 'Departure Delay in Minutes' and 'Arrival Delay in Minutes' to deal with nan values
plt.scatter(new_df['Departure Delay in Minutes'], new_df['Arrival Delay in Minutes'], alpha=0.5)
plt.xlabel('Departure Delay in Minutes')
plt.ylabel('Arrival Delay in Minutes')
plt.title('Departure Delay vs Arrival Delay')
plt.show()
```

```python
# Fill missing values in 'Arrival Delay in Minutes' with 'Departure Delay in Minutes'
mask = new_df['Arrival Delay in Minutes'].isna()
new_df.loc[mask, 'Arrival Delay in Minutes'] = new_df.loc[mask, 'Departure Delay in Minutes']
```

```python
# drop columns with nan values
new_df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
```

```python
# Reorder columns for better readability
sort_col = new_df[['id', 'Gender', 'Age', 'Type of Travel', 'Class', 'Customer Type']]
other_col = [col for col in new_df.columns if col not in sort_col]
new_df = pd.concat([sort_col, new_df[other_col]], axis=1)
```

```python
# Convert column names to lowercase and replace hyphens with spaces
new_df.columns = new_df.columns.str.lower().str.replace('-', ' ')
```

```python
# convert type of column from int to float
new_df['departure delay in minutes'] = new_df['departure delay in minutes'].astype(float)
```
